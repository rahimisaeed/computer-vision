```python
def sketch(image):
    
    image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    image_gray_blur = cv2.GaussianBlur(image_gray, (5, 5), 0)
    
    canny_edges = cv2.Canny(image_gray_blur, 10, 70)
    
    success, mask = cv2.threshold(canny_edges, 70, 255, cv2.THRESH_BINARY_INV)
    return mask
    
video_capture = cv2.VideoCapture(0)

while True:
    
    success, frame = video_capture.read()
    
    cv2.imshow('sketch', sketch(frame))
    if cv2.waitKey(1) == 13:
        break
        
        
video_capture.release()
cv2.destroyAllWindows()
    
    
    
```

```python
import pafy
import cv2

url = 'https://youtu.be/EFEmTsfFL5A'
video = pafy.new(url)
best = video.getbest(preftype="mp4")

capture = cv2.VideoCapture()
capture.open(best.url)

while True:
    ret, frame = capture.read()
    if ret == True:
        cv2.imshow('src', frame)
        if cv2.waitKey(1) == 13:
            break
            
            
capture.release()
cv2.destroyAllWindows()

```
